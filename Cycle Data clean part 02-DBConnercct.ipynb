# Creating an empty Pandas DataFrame, then filling it

```python
#create a new data frame
cols = ['Date', 'Bike_rented', 'Mean_Temperature_C', 'Max_Temperature_F', 'Min_TemperatureF', 'Max_Humidity', 'Min_Humidity',  'Visibility_Miles', 'Wind_Speed_MPH', 'Precipitation_In',
    'Events_num', 'Events', 'month', 'year']
df3 = pd.DataFrame(columns=cols)
```

```python
df3.tail()
```

```python
df3.info()
```

```python
import pandas as pd
Inplace = True
for index, row in df1.iterrows():
    a = row.Date
    kk = row.Mean_Temperature_C
    c = row.Max_Temperature_F
    d = row.Min_TemperatureF
    e = row.Max_Humidity
    f = row.Min_Humidity
    g = row.Mean_Visibility_Miles 
    h1 = row.Mean_Wind_Speed_MPH
    h12 = row.Precipitation_In 
    h2 = row.Events_num 
    h3 = row.Events
    h4 = row.month
    h5 = row.year
    b = 0
    for i, r in df2.iterrows():
        if (row.Date == r.Date):
            b = b + 1
            if not (row.Date == r.Date):
                break
            df3.at[index, 'Date'] = a
            df3.at[index, 'Bike_rented'] = b
            df3.at[index, 'Mean_Temperature_C'] = kk
            df3.at[index, 'Max_Temperature_F'] = c
            df3.at[index, 'Min_TemperatureF'] = d
            df3.at[index, 'Max_Humidity'] = e
            df3.at[index, 'Min_Humidity'] = f
            df3.at[index, 'Visibility_Miles'] = g
            df3.at[index, 'Wind_Speed_MPH'] = h1
            df3.at[index, 'Precipitation_In'] = h12
            df3.at[index, 'Events_num'] = h2
            df3.at[index, 'Events'] = h3
            df3.at[index, 'month'] = h4
            df3.at[index, 'year'] = h5
    print (index,'Date=',a,'Bike_rented',b,c,d,e,f,g,kk)
```
