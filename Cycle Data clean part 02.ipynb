# Data Merging Process

```python
#create a new data frame
cols = ['Date', 
        'Bike_rented', 
        'Max_Temperature_F',                            
        'Min_TemperatureF',                              
        'Max_Humidity',                                     
        'Min_Humidity',                                   
        'Visibility_Miles',                               
        'Wind_Speed_MPH',                         
        'Precipitation_In',        
        'Events',                        
        'Temperature_C',             
        'Events_num',                     
        'month',                           
        'year']
df3 = pd.DataFrame(columns=cols)
```

```python
df3.info()
```

```python
df3.head()
```

```python
import pandas as pd
Inplace = True
for index, row in df1.iterrows():
    if row.notnull:
        a1 = row.Max_Temperature_F               
        a2 = row.Min_TemperatureF                                
        a3 = row.Max_Humidity                                       
        a4 = row.Min_Humidity                            
        a5 = row.Mean_Visibility_Miles                    
        a6 = row.Mean_Wind_Speed_MPH                           
        a7 = row.Precipitation_In              
        a8 = row.Events                         
        a9 = row.Mean_Temperature_C              
        a10 = row.Events_num                      
        a11 = row.month                           
        a12 = row.year
        a13 = row.Date
        b = 0
        for i, r in df2.iterrows():
            if (row.Date == r.Date):
                b = b + 1
                if not (row.Date == r.Date):
                    break
                    df3.at[index, 'Max_Temperature_F'] = a1                              
                    df3.at[index, 'Min_TemperatureF'] = a2                                  
                    df3.at[index, 'Max_Humidity'] = a3                                        
                    df3.at[index, 'Min_Humidity'] = a4                             
                    df3.at[index, 'Visibility_Miles'] = a5 
                    df3.at[index, 'Wind_Speed_MPH'] = a6                         
                    df3.at[index, 'Precipitation_In'] = a7           
                    df3.at[index, 'Events'] = a8                         
                    df3.at[index, 'Mean_Temperature_C'] = a9              
                    df3.at[index, 'Events_num'] = a10                     
                    df3.at[index, 'month'] = a11                           
                    df3.at[index, 'year'] = a12
                    df3.at[index, 'Date'] = a13
                    df3.at[index, 'Bike_rented'] = b
    print (index,'vlaue of rented bike',b,'Dated',a13)
```
