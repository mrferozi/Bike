```python
import pandas as pd
Inplace = True
for index, row in df1.iterrows():
    a = row.Date
    c = row.Mean_Temperature_C 
    d = row.Mean_Humidity
    e = row.Mean_Visibility_Miles 
    f = row.Mean_Wind_Speed_MPH
    g = row.Events_num 
    b = 0
    for i, r in df2.iterrows():
        if (row.Date == r.Date):
            b = b + 1
            if not (row.Date == r.Date):
                break
            print (index,a,b,c,d,e,f,g)
            df3.at[index, 'Date'] = a
            df3.at[index, 'Bike_rented'] = b
            df3.at[index, 'Temp'] = c
            df3.at[index, 'Humidity'] = d
            df3.at[index, 'Visibility_Miles'] = e
            df3.at[index, 'Wind_Speed_MPH'] = f
            df3.at[index, 'Events_num'] = g
```
